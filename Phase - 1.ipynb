```python
driver = webdriver.Chrome("chromedriver.exe")
```

```python
driver.get("https://www.indiacensus.net/literacy-rate.php")
```

```python
data = driver.find_elements(By.TAG_NAME,"td")
```

```python
Data = []
for i in data:
    Data.append(i.text.replace('%','').replace(',',''))
    
len(Data)
```

```python
Data[:10]
```

```python
state_name = Data[1:181:5]
state_name[:10]
```

```python
population =[] 
for i in Data[2:181:5]:
    population.append(int(i))
```

```python
literate_people = []
for i in Data[3:181:5]:
    literate_people.append(int(i))
```

```python
literacy_rate =  []
for i in Data[4:181:5]:
    literacy_rate.append(float(i))
```

```python
len(state_name),len(population),len(literate_people),len(literacy_rate)
```

```python
state_info_df = pd.DataFrame()
```

```python
state_info_df["State name"] = state_name
state_info_df["population"] = population
state_info_df["Literate population"] = literate_people
state_info_df["Literacy rate"] = literacy_rate

'''sourse = ndiacensus.net/literacy-rate.php'''
state_info_df.head()
```
