```python
final_df['Total Crimes'] = final_df.sum(axis=1)
```

```python
# importing Clustring algorithum
from sklearn.cluster import KMeans
```

```python
x = final_df.drop(['DISTRICT','STATE/UT','Total Crimes'],axis=1)
x.head()
```

```python
#Elbow Method for finding Best Number Clusters
wcss = []
for i in range(1,10):
    k_mean = KMeans(n_clusters=i,random_state=18)
    k_mean.fit(x)
    wcss.append(k_mean.inertia_)
```

```python
plt.plot(range(1,10),wcss)
plt.title("Elbow Method")
plt.xlabel("Number of Clusters")
plt.ylabel("WCSS")
plt.show()
```

```python
kmeans = KMeans(n_clusters=3,random_state=18)
pred_kmean = kmeans.fit_predict(x)
```

```python
from sklearn.metrics import silhouette_score
```

```python
silhouette_score(x,pred_kmean)
```
